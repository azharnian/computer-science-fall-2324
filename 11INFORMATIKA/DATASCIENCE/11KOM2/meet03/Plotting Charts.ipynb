## Plotting Charts
using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
year = list(range(2009, 2019))
christian = [77, 76, 75, 73, 73, 71, 69, 68, 67, 65]
unaffiliated = [17, 17, 19, 19, 20, 21, 24, 23, 25, 26]
```

```python
plt.plot(year, christian, label="christian")
plt.plot(year, unaffiliated, label="unaffiliated")

plt.ylim(0, 80)
plt.xlim(2008, 2019)

#decorating
plt.title("religious affiliated of U.S. adults".title())
plt.xlabel("year")
plt.ylabel("% of adults")
plt.legend()
plt.show()
```

```python
name_list = [
    'Lobster roll',
    'Chicken caesar',
    'Bang bang chicken',
    'Ham and cheese',
    'Tuna and cucumber',
    'Egg'
]

boston_price_list = [9.99, 7.99, 7.49, 7, 6.29, 4.99]
london_price_list = [7.5, 5, 4.4, 5, 3.75, 2.25]

plt.hlines(name_list, boston_price_list, london_price_list, color="gray")
plt.plot(boston_price_list, name_list, marker="o", linestyle="", label="boston")
plt.plot(london_price_list, name_list, marker="s", linestyle="", label="london")
```
