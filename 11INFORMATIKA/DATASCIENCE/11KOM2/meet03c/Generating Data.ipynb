```python
import matplotlib.pyplot as plt
```

```python
x_values = list(range(100))
squares = [x**2 for x in range(100)]
cubes = [x**3 for x in range(100)]
rects = [x*3 for x in range(100)]
circles = [3.14*x**2 for x in range(100)]

plt.subplot(2, 2, 1)
plt.plot(x_values, squares)
plt.title("Square Number")

plt.subplot(2, 2, 2)
plt.plot(x_values, cubes)
plt.title("Cubic Number")

plt.tight_layout()
plt.show()
```

## Using Subplots

```python
fig, ax = plt.subplots()
# ax.plot(squares)
ax.scatter(x_values, squares, s=3, c=squares, cmap=plt.cm.Reds)
ax.set_xlabel("numbers")
ax.set_ylabel("values of squares")
ax.set_title("Square Number")
plt.show()
```
