```python
import pandas as pd

file = "train_and_test2.csv"

passengers = pd.read_csv(file)
passengers.head()
```

```python
passengers.columns
```

```python
passengers_clean = passengers[["Passengerid", "Age", "Fare", "Sex", "Pclass"]]
passengers_clean.head()
```

```python
sex = passengers_clean["Sex"].value_counts()
sex
```

```python
import matplotlib.pyplot as plt
fares = passengers_clean["Fare"].value_counts().sort_index()
fares.hist(bins=10)
plt.xlabel("Fare")
plt.ylabel("Number of passengers")
plt.title("Fare Variant in Titanic")
plt.show()
```
