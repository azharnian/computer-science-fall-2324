## Zipf Law Implementation

```python
import matplotlib.pyplot as plt
```

```python

filename = "pg2600.txt"
all_punctuation = ',.-:[#]*/“!?”;()%@'
first_line = "CHAPTER I\n"
last_line = "*** END OF THE PROJECT GUTENBERG EBOOK WAR AND PEACE ***\n"

with open(filename, "r", encoding="utf-8") as file:
    lines = file.readlines()

# print(lines[0:5])
unique_words = {}
start = False
for line in lines:
    if line == first_line:
        start = True
    if line == last_line:
        break
    if start:
        words = line.split()
        for word in words:
            word = word.lower()
            word = word.strip(all_punctuation)
            if word in unique_words:
                unique_words[word] += 1
            else:
                unique_words[word] = 1
# print(unique_words)
freqs = list(unique_words.values())
freqs.sort(reverse=True)
# print(freqs)
ranks = list(range(1,len(freqs)+1))
plt.plot(ranks, freqs, label="Freqs word in 'War and Peace' Book")
plt.xscale("log")
plt.yscale("log")
xi, xf = ranks[0], ranks[-1]
yi, yf = freqs[0], freqs[-1]
plt.plot((xi, xf), (yi, yf), color="red", label="simple regression")
plt.legend()
plt.grid(True) 
plt.show()
```

```python
sorted_unique_words = dict(sorted(unique_words.items(), key = lambda item : item[1], reverse=True))
# print(sorted_unique_words)
top_10 = list(sorted_unique_words.keys())[:10]
freqs_top_10 = list(sorted_unique_words.values())[:10]
print(top_10)
print(freqs_top_10)
```
