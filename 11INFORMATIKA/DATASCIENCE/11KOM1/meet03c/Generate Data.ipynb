```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1, 6))
squares = [n**2 for n in x]
print(squares)
```

```python
cubes = [n**3 for n in x]
print(cubes)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, squares, label="squares")
plt.title("Square Number")

plt.subplot(2, 2, 2)
plt.plot(x, cubes, label="cubes")
plt.title("Cubes Number")

randoms = list(range(10, 51, 10))
plt.subplot(2, 2, 3)
plt.plot(x, randoms, label="randoms")
plt.title("Random Number")


plt.tight_layout()
plt.show()
```

## Using Subplots, Fig, Axes

```python
squares = [x**2 for x in range(1, 1000)]

fig, axes = plt.subplots() # plt.plot()
# axes.plot(squares)
x = list(range(1, 1000))
# axes.plot(x, squares, linewidth=3, marker="o", linestyle="")

# axes.scatter(x, squares, s=5, c="red")
axes.scatter(x, squares, s=5, c=squares, cmap=plt.cm.Reds)
axes.set_title("square numbers".title(), fontsize=40)
axes.set_xlabel("values", fontsize=30)
axes.set_ylabel("square of value", fontsize=30)

plt.show()
```
