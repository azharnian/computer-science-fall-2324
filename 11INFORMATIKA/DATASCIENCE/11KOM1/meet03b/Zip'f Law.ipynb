## Zip'f Law

```python
import matplotlib.pyplot as plt

book_file = "pg2600.txt"
first_line = "CHAPTER I\n"
last_line = "*** END OF THE PROJECT GUTENBERG EBOOK WAR AND PEACE ***\n"
all_punctuation = ',.-:[#]*/“!?”;()%@—’'

with open(book_file , "r", encoding="utf-8") as f:
    lines = f.readlines()

unique_words = {}
start = False
for line in lines:
    if line == first_line:
        start = True
    
    if line == last_line:
        break
    
    if start:
        words = line.split()

        for word in words:
            word = word.lower()
            word = word.strip(all_punctuation)
            if word in unique_words:
                unique_words[word] += 1
            else:
                unique_words[word] = 1
    
# print(unique_words)
freqs = list(unique_words.values())
freqs.sort(reverse=True)
# print(freqs)
ranks = list(range(1, len(freqs)+1))
# print(ranks)

plt.plot(ranks, freqs)
# plt.gca().invert_xaxis()
plt.xscale("log")
plt.yscale("log")

xi, xf = ranks[0], ranks[-1]
yi, yf = freqs[0], freqs[-1]
plt.plot([xi, xf], [yi, yf], color="gray")
```

```python
sorted_unique_words = dict(sorted(unique_words.items(), key = lambda item : item[1], reverse = True))
# print(list(sorted_unique_words.keys()))
# print(list(sorted_unique_words.values()))
freqs = list(sorted_unique_words.values())[:5]
ranks = list(sorted_unique_words.keys())[:5]
plt.plot(ranks, freqs)
```
