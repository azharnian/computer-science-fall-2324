```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
# subplot(nrows, ncols, index)
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Sin(x)')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Cos(x)')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Tan(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Exp(x)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
